```python
asphalt_unit_cost = 0
asphalt_quantity=0
concrete_unit_cost = 0
concrete_quantity = 0
labor_unit_cost = 0
labor_hours = 0
```

```python
asphalt_unit_cost = input('Enter asphalt_unit_cost: ')
asphalt_quantity=input('Enter asphalt_quantity: ')
concrete_unit_cost = input('Enter concrete_unit_cost: ')
concrete_quantity = input('Enter concrete_quantity: ')
labor_unit_cost = input('Enter labor_unit_cost: ')
labor_hours = input('Enter labor_hours: ')
```

```python
asphalt_total_cost = int(asphalt_unit_cost)*int(asphalt_quantity)
concrete_total_cost = int(concrete_unit_cost)*int(concrete_quantity)
labor_total_cost = int(labor_unit_cost)*int(labor_hours)
total_cost = asphalt_total_cost+concrete_total_cost+labor_total_cost
print(asphalt_total_cost)
print(concrete_total_cost)
print(labor_total_cost)
print(total_cost)

threshhold = 50000
if total_cost<50000:
  print("Project is under budget")
else:
  print("Project is over budget")
```
