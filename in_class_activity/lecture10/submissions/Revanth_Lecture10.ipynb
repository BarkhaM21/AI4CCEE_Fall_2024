```python
# Step 1: Import the required module
from tensorflow.keras.datasets import cifar10

```

```python
# Step 2: Load the CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()


```

```python
# Step 3: Normalize pixel values to the range 0 to 1
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# Check the shape and range of data
print(f"Train Images Shape: {train_images.shape}")
print(f"Train Labels Shape: {train_labels.shape}")
print(f"Test Images Shape: {test_images.shape}")
print(f"Test Labels Shape: {test_labels.shape}")
print(f"Pixel Value Range: Min = {train_images.min()}, Max = {train_images.max()}")

```

```python
# Step 4: Verify the normalization and display some example images
import matplotlib.pyplot as plt

# Display 10 sample images from the training set
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i][0]])
plt.tight_layout()
plt.show()
```

```python
# Step 6: Divide both train_images and test_images by 255.0
train_images = train_images / 255.0
test_images = test_images / 255.0


```

```python
# Step 7: Verify the download process
print("Download Verification:")
print(f"Training data shape: {train_images.shape}, Training labels shape: {train_labels.shape}")
print(f"Testing data shape: {test_images.shape}, Testing labels shape: {test_labels.shape}")

```

```python
# Step 8: Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Step 9: Create a list named class_names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Step 10: Use plt.figure() to create a figure for the plot
plt.figure(figsize=(10, 10))

# Step 11: Loop through the first 25 images
for i in range(25):
    # Step 12: Create a subplot for each image
    plt.subplot(5, 5, i + 1)  # 5 rows, 5 columns, index = i + 1
    plt.xticks([])  # Remove x-axis ticks
    plt.yticks([])  # Remove y-axis ticks
    plt.grid(False) # Remove grid
    plt.imshow(train_images[i])  # Display the image
    plt.xlabel(class_names[train_labels[i][0]])  # Set label as class name

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Step 10: Use plt.figure() to create a figure for the plot
plt.figure(figsize=(10, 10))  # Create a figure with size 10x10

# Step 11–16: Loop through the first 25 images
for i in range(25):
    plt.subplot(5, 5, i + 1)  # 5 rows, 5 columns, index = i + 1
    plt.xticks([])  # Step 13: Remove x-axis ticks
    plt.yticks([])  # Step 13: Remove y-axis ticks
    plt.grid(False)  # Step 13: Disable the grid
    plt.imshow(train_images[i])  # Step 14: Display the image
    plt.xlabel(class_names[train_labels[i][0]])  # Step 15: Add a label using class name

# Step 16: Show the plot
plt.tight_layout()  # Adjust layout to avoid overlapping
plt.show()
```

```python
# Step 17: Import the necessary modules
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D

# Step 18: Define a sequential model
model = Sequential()

# Step 19: Add the first convolutional layer
model.add(Conv2D(
    filters=32,              # Number of filters
    kernel_size=(3, 3),      # Dimensions of the filter
    activation='relu',       # Activation function
    input_shape=(32, 32, 3)  # Input shape: (height, width, channels)
))

# Display the model's architecture
model.summary()
```

```python
from tensorflow.keras.layers import Flatten, Dense

# Step 20: Print model summary
print("Initial Model Architecture:")
model.summary()


```

```python
# Step 21: Add a Flatten layer
model.add(Flatten())  # Flattens the feature maps into a 1D vector

```

```python
# Step 22: Add a Dense layer with 64 units
model.add(Dense(
    units=64,         # Number of neurons
    activation='relu' # Activation function
))

```

```python
# Step 23: Add an output Dense layer
model.add(Dense(
    units=10  # Number of output units for 10 classes
))  # No activation function specified (logits are directly output)

```

```python
# Step 24: Compile the model
from tensorflow.keras.losses import SparseCategoricalCrossentropy

model.compile(
    optimizer='adam',  # Adaptive gradient optimization
    loss=SparseCategoricalCrossentropy(from_logits=True),  # Loss function with logits
    metrics=['accuracy']  # Track accuracy during training
)

# Print the final model summary
print("\nFinal Model Architecture:")
model.summary()
```

```python
# Step 25: Train the model
history = model.fit(
    train_images, train_labels,              # Training data
    epochs=10,                               # Number of epochs
    validation_data=(test_images, test_labels)  # Validation data
)


```

```python
# Step 26: Plot training and validation accuracy
import matplotlib.pyplot as plt

# Extract accuracy and validation accuracy from history
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Plot accuracy trends
plt.figure(figsize=(8, 6))
plt.plot(train_accuracy, label='Training Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.xlabel('Epoch')  # X-axis label
plt.ylabel('Accuracy')  # Y-axis label
plt.ylim([0.5, 1])  # Set y-axis limits
plt.legend(loc='lower right')  # Add legend
plt.title('Training and Validation Accuracy')
plt.show()

```

```python
# Step 27: Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_images, test_labels, verbose=2)

print(f"\nTest Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```
