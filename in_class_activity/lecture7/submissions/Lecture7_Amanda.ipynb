```python
#tep 3 Find how many unique journey_ids in the data filtered as per (1) (Ans: 7)
SELECT COUNT(DISTINCT journeyid) FROM gps_data
WHERE month = 10 AND day = 3 AND hour = 4;

#Step 4 Filter data on Oct 1st and 3rd and hour between 3AM and 6AM (101116 rows)
SELECT * FROM gps_data
WHERE month = 10 AND
day IN (1, 3) AND
hour BETWEEN 3 AND 6;

#Step 5 Select all journey_ids contain abc from whole gps_data table (find 8 of them)
SELECT distinct journeyid FROM gps_data
WHERE journeyid LIKE '%abc%';

#Step 6 Count how many datapoint_ids start with 9 (Ans: 15256)
SELECT COUNT(*) FROM gps_data
WHERE datapointid LIKE '9%';

#Step 7 Select all rows where journey_id ends with cc (66)
SELECT * FROM gps_data
WHERE journeyid LIKE '%cc';

#Step 8 Select all columns, also prepare a new column that combines datapoint_id and day with a - symbol
SELECT *, CONCAT(datapointid, '-', day) AS datapoint_day
FROM gps_data;

```

```python
#Activity 3: Aggregations
#Step 1 Find maximum speed from table vehicle_data (Ans: 162.43)
SELECT MAX(speed) AS max_speed
FROM vehicle_data;

#Step 2 Find average, minimum, maximum and standard deviations of speeds for CHEVROLET make
SELECT
    AVG(speed) AS avg_speed,
    MIN(speed) AS min_speed,
    MAX(speed) AS max_speed,
    STDDEV(speed) AS stddev_speed
FROM vehicle_data
WHERE make = 'CHEVROLET';

#Step 3 Create seg_id by combining route_id and segment_start_measure
SELECT *, CONCAT(route_id, '-', segment_start_measure) AS seg_id
FROM vehicle_data;

#Step 4 Add this seg_id to existing vehicle_data table and verify it  Alter_table and update
ALTER TABLE vehicle_data
ADD COLUMN seg_id VARCHAR(255);

UPDATE vehicle_data
SET seg_id = CONCAT(route_id, '-', segment_start_measure);

#Verify by selecting the updated table
SELECT * FROM vehicle_data
LIMIT 10;  -- to preview the first 10 rows

#Step 5 Find 85th percentiles speeds for each segment_id
SELECT
    seg_id,
    PERCENTILE_CONT(0.85) WITHIN GROUP (ORDER BY speed) AS speed_85th_percentile
FROM vehicle_data
GROUP BY seg_id;

#Step 6: Find the highest speeds for each model. (Compare 95th percentile speeds for each model)
SELECT
    model,
    PERCENTILE_CONT(0.95) WITHIN GROUP (ORDER BY speed) AS speed_95th_percentile
FROM vehicle_data
GROUP BY model
ORDER BY speed_95th_percentile DESC;

```

```python
#Activity 4: Joins
#Step 1 Join the gps_data and vehicle_data tables using JOIN on datapoint_id
SELECT * FROM gps_data
JOIN vehicle_data ON gps_data.datapointid = vehicle_data.datapointid;

#Step 2 Select datapoint_id, journey_id, speed, seg_id from the above join
SELECT
    gps_data.datapointid,
    gps_data.journeyid,
    vehicle_data.speed,
    vehicle_data.seg_id
FROM gps_data
JOIN vehicle_data ON gps_data.datapointid = vehicle_data.datapointid;

#Step 3 Count the unique journey_ids that came to stop at least once (Speed = 0)
SELECT COUNT(DISTINCT gps_data.journeyid) AS stopped_journey_count
FROM gps_data
JOIN vehicle_data ON gps_data.datapointid = vehicle_data.datapointid
WHERE vehicle_data.speed = 0;

```

```python
#Activity 5: Nested queries
#Step 1. Filter points in gps_data with model = GMC from vehicle_data
#1. Use join
SELECT gps_data.*
FROM gps_data
JOIN vehicle_data ON gps_data.datapointid = vehicle_data.datapointid
WHERE vehicle_data.model = 'GMC';

#2. Use nested query
SELECT *
FROM gps_data
WHERE datapointid IN (
    SELECT datapointid
    FROM vehicle_data
    WHERE model = 'GMC'
);

#Step 2. Count number of unique journeys in each car make and model using nested query order by make and model in descending order for each
SELECT make, model, COUNT(DISTINCT gps_data.journeyid) AS unique_journeys
FROM vehicle_data
JOIN gps_data ON gps_data.datapointid = vehicle_data.datapointid
GROUP BY make, model
ORDER BY make DESC, model DESC;

#Step 3. Filter the gps data of the journey with the highest speed from the whole dataset
#i. You should find the jounrney_id with max speed from joining tables (JID: a37cf85fe887cb5b54fd828664ace413938a6e5d)
SELECT gps_data.journeyid
FROM gps_data
JOIN vehicle_data ON gps_data.datapointid = vehicle_data.datapointid
ORDER BY vehicle_data.speed DESC
LIMIT 1;

#ii. Then filter the corresponding journey data from gps_data (Ans: 343 rows)
SELECT *
FROM gps_data
WHERE journeyid = 'a37cf85fe887cb5b54fd828664ace413938a6e5d';

```

```python
#Activity 6
#Step 1. Find the distance between points below:
# Point 1: 41.948637, -93.610065, Point 2: 41.997339, -93.610229
SELECT ST_Distance_Sphere(
    ST_SetSRID(ST_Point(-93.610065, 41.948637), 4326),
    ST_SetSRID(ST_Point(-93.610229, 41.997339), 4326)
) AS distance_meters;

#Step 2. Create a line string from the above points and calculate the distance from Point 3 : (41.972017, -93.619738)
SELECT ST_Distance(
    ST_MakeLine(
        ST_SetSRID(ST_Point(-93.610065, 41.948637), 4326),
        ST_SetSRID(ST_Point(-93.610229, 41.997339), 4326)
    ),
    ST_SetSRID(ST_Point(-93.619738, 41.972017), 4326)
) AS distance_from_point3_meters;

#Step 3. Make point geometry from latitude and longitude in the gps_data table
ALTER TABLE gps_data
ADD COLUMN geom GEOMETRY(Point, 4326);

UPDATE gps_data
SET geom = ST_SetSRID(ST_Point(longitude, latitude), 4326);

#Step 4. Create a buffer around point: 3 for 3 miles and filter all datapoints from gps_data within the buffer
SELECT ST_Buffer(
    ST_Transform(ST_SetSRID(ST_Point(-93.619738, 41.972017), 4326), 3857),
    4828.03  -- 3 miles in meters (approximately 1609.34 meters per mile)
) AS buffer_geom;

SELECT *
FROM gps_data
WHERE ST_Within(
    geom,
    ST_Transform(
        ST_Buffer(ST_Transform(ST_SetSRID(ST_Point(-93.619738, 41.972017), 4326), 3857), 4828.03),
        4326
    )
);
```
